```python
import numpy as np
import pandas as pd
```

```python
list1=[1,2,3,4,5]
list2=[10,20,30,40,50]
array1 = np.array(list1)
array2 = np.array(list2)

print("list1:", type(list1))
print("array1:", type(array1))
```

```python
#multiply
multiplication= array1 * array2
print("multiplication is:",multiplication)
```

```python
#divion
division= array1 / array2
print("division is:",division)
```

```python
#poweroff
poweroff=array1 ** 2
print("poweroff of array1 is:",poweroff)

poweroff=array2 ** 3
print("poweroff of array2 is:",poweroff)
```

```python
#addition of two
addition= array1 + array2
print("addition of two:",addition)
```

```python
sin_value=np.sin(array1)
log_value=np.log(array2)
log2_value=np.log2(array1)
exp_value=np.exp(array2)

print("sin value:", sin_value)
print("log value:",log_value)
print("log2 value:",log2_value)
print("exp value:",exp_value)
```

```python
import matplotlib.pyplot as plt
```

```python
# Given information
people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]
```

```python
#scatter plot
plt.figure(figsize=(8,6))
plt.scatter(age, height,s = weight,c='blue',alpha=0.7)

# Adding labels and title
plt.title('Scatter Plot - Age vs Height (Weight represented by size)')
plt.xlabel('Age')
plt.ylabel('Height')
# Show the plot
plt.show()
```

```python
data = {'PEOPLE': ['Kiran', 'Arun', 'Vijay', 'Varun'],
        'AGE': [25, 30, 35, 40],
        'HEIGHT': [145, 151, 165, 173],
        'WEIGHT': [45, 55, 65, 75]}
```

```python
df = pd.DataFrame(data)
```

```python
# Histogram for Age
plt.hist(df['AGE'], bins=10, color='blue', edgecolor='black')
plt.title('Histogram of Age for People')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Histogram for Height
plt.hist(df['HEIGHT'], bins=10, color='green', edgecolor='black')
plt.title('Histogram of Height for People')
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.show()

# Histogram for Weight
plt.hist(df['WEIGHT'], bins=10, color='orange', edgecolor='black')
plt.title('Histogram of Weight for People')
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.show()
```
